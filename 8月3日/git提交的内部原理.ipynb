关键留意“object”和“refs”这两个目录

```powershell
\.GIT
├───hooks
├───info
├───objects【存储所有Git对象（blob/tree/commit）】
│   ├───e6
│   ├───info
│   └───pack
└───refs
    ├───heads【指针】
    └───tags【标签】
```

<link rel='stylesheet' href='../style.css'>

1. **objects与refs**
   - `objects/`：存储所有Git对象（blob/tree/commit），通过SHA-1哈希寻址
   - `refs/`：保存分支指针（heads）和标签指针（tags）

2. **区域划分**
   - 暂存区：`.git/index`（二进制文件，存储准备提交的快照）
   - 代码仓库：`.git/objects/`（实际存储所有版本数据的数据库）

3. **指针机制**
   - HEAD指针：`.git/HEAD` 文件（总是指向当前分支）
   - 分支指针：`.git/refs/heads/` 下的文件（如dev）
   - 工作原理：指针本质是包含SHA-1值的文本文件，通过引用跳转实现版本切换

4. `git commit -m` 执行流程：
   - ① 冻结index内容为tree对象
   - ② 创建commit对象（包含作者/提交者信息、tree指针、父commit指针）
   - ③ 更新当前分支指针（如refs/heads/dev）
   - ④ 移动HEAD指针到新commit

（可通过 `git cat-file -p <hash>` 查看具体对象内容，验证上述机制）
