### 学会看查 Git Graph

<font color='dodgerblue'><b>蓝色线条</b></font>：本地分支（如 `main`）的提交历史。

<font color='deeppink'><b>粉色线条</b></font>：远程分支（如 `origin/main`）的提交历史。

   - **`main`**：这是本地分支，显示为蓝色，对应的提交信息是 `v2`。







   - **`origin/HEAD` 和 `origin/main`**：这两个是远程分支，显示为粉色，其中 `origin/HEAD` 是远程仓库的默认分支（通常是 `origin/main`）的别名。
